## Explore the “Crop” data:

###  10. What is the total occurrence (count used) of each crop type in the sample?

```python
# Count the occurrence of each crop type
crop_type_counts = data['Crop Type'].value_counts()

# Display the count of each crop type
print(crop_type_counts)

# Plot the total occurence for each crop type

crop_type_counts.plot(kind='bar', color=['blue', 'yellow', 'orange'])
plt.xlabel('Crop Type')
plt.ylabel('Total count')
plt.title('Total occurence of each Crop Type')
plt.show()
```

### 11. What is the break-down (count used) of “crop type” at each location?

```python
# Count the occurrence of each crop type at each location
croptype_count = data.groupby(['Location', 'Crop Type']).size()

# Display the breakdown of crop type at each location
print(croptype_count)
```

```python
# Group the data by 'Location' and 'Crop Type' and get the count
breakdown = data.groupby(['Location', 'Crop Type']).size().reset_index(name='Count')

# Pivot the data to have crop types as columns
pivot_df = breakdown.pivot(index='Location', columns='Crop Type', values='Count')

# Plot the data as a bar plot
pivot_df.plot(kind='bar')

# Set the labels and title
plt.xlabel('Location')
plt.ylabel('Count')
plt.title('Breakdown of Crop Types at Each Location')

# Display the plot
plt.show()
```

### 12. What is the total yield for each crop type over the entire sample?

```python
# Calculate total yield for each crop type
total_yield = data.groupby('Crop Type')['Yield'].sum()

print(total_yield)

# Plot the total yield for each crop type

total_yield.plot(kind='bar', color=['blue', 'yellow', 'orange'])
plt.xlabel('Crop Type')
plt.ylabel('Total Yield')
plt.title('Total Yield for Each Crop Type')
plt.show()
```

### Oats is having highest crop yield

### 13. Has there been any changes in crop type usage over time?

```python
# Group the data by 'Year' and 'Crop Type' and calculate the count
crop_usage = data.groupby(['Year', 'Crop Type']).size().reset_index(name='Count')

# Create the grouped bar plot
sns.barplot(data=crop_usage, x='Year', y='Count', hue='Crop Type', dodge=False)

# Set the labels and title
plt.xlabel('Year')
plt.ylabel('Crop Type Usage')
plt.title('Changes in Crop Type Usage over Time')

# Rotate the x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

### From the above plot it is evident that oats usage have increased a lot compared to wheat and Maize

### 14. Who was the longest serving staff member(s) making observations?
### How many years and at what location?

```python
# Group the data by observer
observer_grouped = data.groupby('Observer')

# Calculate the duration of service for each observer
observer_duration = observer_grouped['Year'].apply(lambda x: x.max() - x.min() + 1)  # Add 1 to include the minimum year

# Find the maximum duration
max_duration = observer_duration.max()

# Get the observer(s) with the longest duration
longest_serving_observers = observer_duration[observer_duration == max_duration]

# Print the results
for observer in longest_serving_observers.index:
    years = (data[data['Observer'] == observer]['Year'].min(), data[data['Observer'] == observer]['Year'].max())
    year_count = data[data['Observer'] == observer]['Year'].nunique()  # Count the unique years
    
    locations = data[data['Observer'] == observer]['Location'].unique()
    
    print(f"Observer: {observer}")
    print(f"Years: {years} ({year_count} Years)")
    print(f"Locations: {locations}")
    print()
```

### Anthony Branch, David Fowler, Kimberly Lindsey and Vanessa Coleman are longest serving staff members for 5 years at Josephfurt, Williamhaven and Andersonport locations.

### 15. Did any staff member join again after they first performed the role?

```python
# Group the data by observer
observer_grouped = data.groupby('Observer')

# Find the observers who have multiple occurrences
observers_with_multiple_occurrences = observer_grouped.filter(lambda x: len(x) > 1)['Observer'].unique()

if len(observers_with_multiple_occurrences) > 0:
    print("Yes, some staff members joined again after their first role.")
    print("Observers with multiple occurrences:", observers_with_multiple_occurrences)
else:
    print("No, all staff members performed the role only once.")
```

### 19. Is there an observable relationship between temperature and the year?
### For all locations combined?

```python
average_temp_year = data.groupby('Year')['Avg Min Temp *C'].mean()
```

```python
plt.plot(average_temp_year.index, average_temp_year.values)
plt.xlabel('Year')
plt.ylabel('Average Temperature (°C)')
plt.title('Average Temperature Over the Years (All Locations)')
plt.show()
```

### From above plot we can see that the average temperature has gradually increased over the years, the relationship is as the year increased temperature also increased.

### At each location?

```python
# Iterate over each unique location
for location in data['Location'].unique():
    location_data = data[data['Location'] == location]
    plt.plot(location_data['Year'], location_data['Avg Min Temp *C'], label=location)

# Set the plot title and labels
plt.title('Temperature Variation by Year and Location')
plt.xlabel('Year')
plt.ylabel('Temperature (°C)')

# Add a legend to differentiate the locations
plt.legend()

# Display the plot
plt.show()
```

### From above plot, temperature is inceased over the years and Andersonport is having highest temperature among 3 locations.

### 20. Is there an observable relationship between crop yield and the year?
### For all locations combined?

```python
# Calculate the average yield for each year
yield_by_year = data.groupby('Year')['Yield'].mean()

# Plot the average yield over time for all locations combined
plt.figure(figsize=(10, 6))
plt.plot(yield_by_year.index, yield_by_year.values)
plt.title('Crop Yield Trend (All Locations)')
plt.xlabel('Year')
plt.ylabel('Average Yield')
plt.show()
```

### There is no constant relationship between year and yeild, for few years yield incearsed, and for few yeild decreased compared to previous years.

### At each location?

```python
# Calculate the average yield for each location and year
yield_by_location = data.groupby(['Location', 'Year'])['Yield'].mean()

# Plot the average yield over time for each location
locations = data['Location'].unique()
for location in locations:
    yield_location = yield_by_location.loc[location]
    plt.figure(figsize=(10, 6))
    plt.plot(yield_location.index, yield_location.values)
    plt.title(f'Crop Yield Trend ({location})')
    plt.xlabel('Year')
    plt.ylabel('Average Yield')
    plt.show()
```

 ### There is no constant relationship between year and yeild, for few years yield incearsed, and for few yeild decreased compared to previous years.

### 21. Is there an observable relationship between rain data and the year:
### For all locations combined?

```python
import matplotlib.pyplot as plt

# Group the data by 'Year' and calculate the average rain data
rain_year_combined = data.groupby('Year')['Yr Rain mm'].mean()

# Plot the average rain data for all locations combined
plt.plot(rain_year_combined.index, rain_year_combined.values)
plt.xlabel('Year')
plt.ylabel('Average Rainfall (mm)')
plt.title('Average Rainfall over Years (All Locations)')

# Display the plot
plt.show()
```

### From the above plot is clear that rainfall has gradually increased over years

### At each location?

```python
# Get the unique locations
locations = data['Location'].unique()

# Create subplots for each location
fig, axs = plt.subplots(len(locations), 1, figsize=(10, 8), sharex=True)

# Iterate over each location and plot the rain data
for i, location in enumerate(locations):
    location_data = data[data['Location'] == location]
    axs[i].plot(location_data['Year'], location_data['Yr Rain mm'])
    axs[i].set_ylabel('Rainfall (mm)')
    axs[i].set_title(f'Rainfall over Years - {location}')

# Set the x-axis label
axs[-1].set_xlabel('Year')

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

### Rainfall has gradually increased over the years and among 3 locations Josephfurt registeres with highest rainfall of 2030 mm in 2018.
